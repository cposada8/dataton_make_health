# Merge all the data

## read height database

```python
height_df = pd.read_csv('../features/altura.csv', sep=";")
```

```python
height_df
```

```python
df = df.merge(height_df,how='left', left_on='Municipality code', right_on='Municipality code',suffixes=('_left', '_right'))
df
```

## Read population density database

```python
density_df = pd.read_csv('../features/density.csv', sep=";")
density_df
```

```python
# transform the density data
density_data = density_df.melt(id_vars=["Municipality code"], 
        var_name="Date", 
        value_name="density_year")
density_data
```

```python
density_data['year'] = density_data.Date.apply(lambda x: int(x[-4:]))
```

```python
density_data['concat_code_year'] = density_data['Municipality code'].astype('str') + "|" + \
                                    density_data['year'].astype('str')
density_data
```

```python
df = df.merge(density_data[['concat_code_year', 'density_year']],how='left', on='concat_code_year', suffixes=('_left', '_right'))
df
```

```python
df.info()
```

```python
# save the file
df.to_csv("../features/data_colombia_completa.csv", sep=";", index=False)
```
