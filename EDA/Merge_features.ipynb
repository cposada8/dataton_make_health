# Merge all the data

## read height database

```python
height_df = pd.read_csv('../features/altura.csv', sep=";")
```

```python
height_df
```

```python
df = df.merge(height_df,how='left', left_on='Municipality code', right_on='Municipality code',suffixes=('_left', '_right'))
df
```

## Read population density database

```python
density_df = pd.read_csv('../features/density.csv', sep=";")
density_df
```

```python
df = df.merge(density_df,how='left', left_on='Municipality code', right_on='Municipality code',suffixes=('_left', '_right'))
df
```
