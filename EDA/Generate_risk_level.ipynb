```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../features/base_colombia_completa.csv", sep=";")
```

```python
data
```

```python
codes = data["Municipality code"].unique()
risk_df = pd.DataFrame()

tasas = []
```

```python
for code in codes:
    data_actual = data[data["Municipality code"]==code]
    tasa_municipio = data_actual["tasa_dengue_mensual"].mean()
    tasas.append(tasa_municipio)
```

```python
tasas_promedio_df= pd.DataFrame({'code': codes.tolist(),'tasa_dengue_promedio': tasas})
```

```python
tasas_promedio_df["tasa_dengue_promedio_x10000_hab"] = tasas_promedio_df["tasa_dengue_promedio"]*10000
```

```python
tasas_promedio_df
```

```python
tasas_promedio_df.to_csv("tasas_dengue_promedio.csv", sep=";", index=False)
```

```python
data_complete = data.merge(tasas_promedio_df, left_on='Municipality code', right_on='code', how='left')
data_complete
```

```python
data_complete.to_csv("../features/data_colombia_completa_2.csv", sep=";", index=False)
```
