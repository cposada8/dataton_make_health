```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../features/base_colombia_completa.csv", sep=";")
```

```python
data
```

```python
data.columns
```

```python
tasas_promedio_df = data.groupby(['Municipality code'])[['tasa_dengue_mensual', 'tasa_dengue_mensual_x10000_habitantes']].mean().reset_index()
```

```python
tasas_promedio_df.rename(columns={'tasa_dengue_mensual':'promedio_tasa_dengue_mensual', 
                                 'tasa_dengue_mensual_x10000_habitantes':'promedio_tasa_dengue_mensual_x10000_habitantes'},
                        inplace=True)
tasas_promedio_df
```

```python
tasas_promedio_df.to_csv("tasas_dengue_promedio.csv", sep=";", index=False)
```

```python
data_complete = data.merge(tasas_promedio_df, on='Municipality code', how='left')
data_complete
```

```python
data_complete.to_csv("../features/data_colombia_completa_2.csv", sep=";", index=False)
```
